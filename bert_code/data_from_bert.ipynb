```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../bert_data/data_from_bert/L6N_20151024topics_and_probs.txt', header=None)
```

```python
df
```

```python
def flag(string):
    flag = 0
    for item in string:
        if item == '[':
            flag = 1
            return flag
        elif item == ']':
            flag = 2
            return flag
        else:
            continue
    return flag        
```

```python
df['aux'] = df[0].apply(lambda x: flag(x) )
```

```python
df
```

```python
df['join'] = np.nan
```

```python
df.iloc[0]['join'] == 3
```

```python
df.at[0, 'join'] = 3
```

```python
start = 3
for index in range(df.shape[0]):
    if df.iloc[index]['aux'] == 1:
        start += 1
        df.at[index, 'join'] = start
    elif df.iloc[index]['aux'] != 1:
        df.at[index, 'join'] = start
```

```python
df_new = df.groupby('join', sort=False)[0].apply(' '.join).reset_index(name='new_seq')
```

```python
import re
def process(row):
    row = row.replace('0. ', '0 ')
    row = row.replace(' ]', ']')
    row = re.sub(r'[\[\]\(\):,]', r'', row)
    row = re.sub(r'(\s){2,}', r' ', row)
    row_split = row.split(' ')
    return ', '.join(row_split)
```

```python
df_new['new_seq'] = df_new['new_seq'].apply(lambda x: process(x))
```

```python
df_new
```

```python
df_new['new_seq'][1]
```

```python
def refine(x):
    if (x.endswith('.')):
        x = x[:-1]
    return x    
```

```python
df_new['new_seq'] = df_new['new_seq'].apply(lambda x: refine(x))
```
